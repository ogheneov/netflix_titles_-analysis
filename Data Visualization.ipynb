## Data Visualization

- Visual representation of data to discover trends and patterns
- Helps in storytelling and decision-making

### Types

- Categorical: Bar chart, Pie chart, etc
- Numerical: Box plot, Histogram, etc
- Relationship: Scatter plot, Line plot, Heatmap, Pair plot, etc

### Libraries

- Matplotlib
- Seaborn

```python
import pandas as pd

import matplotlib.pyplot as plt
```

- Figure: The entire canvas that holds all subplots
- Axes: A single plot

```python
#Create a figure with a single axes

x = [1, 2, 3]

y = [4, 5, 6]

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Basic Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

plt.show() #to display the plot
```

```python
#create a figure with mutliple subplots

fig, axes = plt.subplots(1, 2) #two plots side by side
axes[0].plot(x, y)
axes[0].set_title("Left Plot")
axes[1].plot(x, y)
axes[1].set_title("Right Plot")

plt.show()
```

```python
fig, axes = plt.subplots(2, 2)

axes[0, 0].plot(x,y)
axes[0, 1].plot(x,y)
axes[1, 0].plot(x,y)
axes[1, 1].plot(x,y)

plt.show()
```

## Exploratory Data Analysis (EDA)

- Univariate Analysis: Examining one variable at a time

```python
df = pd.read_csv(r"C:\Users\Ukpoweh Gift\Downloads\netflix_titles (2).csv")

df.head()
```

```python
df.info()
```

What are the number of netflix titles released per year?

```python
title_per_year = df.groupby("release_year")["title"].count().reset_index()

title_per_year
```

```python
#line plot
plt.plot(title_per_year["release_year"], title_per_year["title"], color="red", marker="o")
plt.title("Number of Netflix Titles Released Per Year")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.show()
```

Bivariate analysis

- Examining the relationship between two variables

```python
df["type"].value_counts()
```

```python
#movies and tv shows released per year

release_year_by_type = df.groupby(["release_year", "type"]).size().unstack(fill_value=0).reset_index()

release_year_by_type
```

```python
df.groupby(["release_year", "type"]).size().unstack(fill_value=0).reset_index()
```

```python
plt.plot(release_year_by_type["release_year"], release_year_by_type["Movie"], label="Movies")
plt.plot(release_year_by_type["release_year"], release_year_by_type["TV Show"], label="TV Shows")
plt.legend()

plt.show()
```
